# Analisis de Red de Tiendas - RetailNow

## Descripcion
Este proyecto analiza los datos de ventas, inventarios y satisfaccion del cliente para la cadena de tiendas minoristas RetailNow. 
Utilizamos Pandas para manipulacion de datos y Numpy para calculos estadisticos y simulaciones.

## Objetivo
Optimizar el rendimiento de las tiendas mediante el analisis detallado de:
- Ventas totales por tienda
- Rotacion de inventarios
- Niveles de satisfaccion del cliente
- Proyecciones de ventas futuras

## 1. Preparar el Entorno - Importar Librerias Necesarias

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print('Librerias importadas exitosamente')
print('Pandas version:', pd.__version__)
print('Numpy version:', np.__version__)
```

## 2. Cargar los Datos - Lectura desde Archivos CSV

```python
# Rutas de los archivos CSV
ruta_ventas = 'workspace\sales.csv'
ruta_inventarios = 'workspace\inventories.csv'
ruta_satisfaccion = 'workspace\satisfaction.csv'

# Cargar los datos en DataFrames
df_ventas = pd.read_csv(ruta_ventas)
df_inventarios = pd.read_csv(ruta_inventarios)
df_satisfaccion = pd.read_csv(ruta_satisfaccion)

print('Datos cargados exitosamente')
```

## 3. Limpiar los Datos - Eliminar Valores Nulos

```python
# Eliminar filas con valores nulos
df_ventas = df_ventas.dropna()
df_inventarios = df_inventarios.dropna()
df_satisfaccion = df_satisfaccion.dropna()

print('\nDatos limpiados exitosamente')
print('Filas en df_ventas:', len(df_ventas))
print('Filas en df_inventarios:', len(df_inventarios))
print('Filas en df_satisfaccion:', len(df_satisfaccion))
```

## 4. Exploracion de Datos - Estructura y Resumen Estadistico

```python
# Mostrar las primeras filas de cada DataFrame
print('PRIMERAS FILAS DE DATOS DE VENTAS:')
print('-' * 50)
print(df_ventas.head())
print('\nESTRUCTURA DEL DATAFRAME DE VENTAS:')
print(df_ventas.info())
print('\nESTADISTICAS DESCRIPTIVAS DE VENTAS:')
print(df_ventas.describe())
```

```python
# Informacion de inventarios
print('PRIMERAS FILAS DE DATOS DE INVENTARIOS:')
print('-' * 50)
print(df_inventarios.head())
print('\nESTRUCTURA DEL DATAFRAME DE INVENTARIOS:')
print(df_inventarios.info())
print('\nESTADISTICAS DESCRIPTIVAS DE INVENTARIOS:')
print(df_inventarios.describe())
```

```python
# Informacion de satisfaccion
print('PRIMERAS FILAS DE DATOS DE SATISFACCION:')
print('-' * 50)
print(df_satisfaccion.head())
print('\nESTRUCTURA DEL DATAFRAME DE SATISFACCION:')
print(df_satisfaccion.info())
print('\nESTADISTICAS DESCRIPTIVAS DE SATISFACCION:')
print(df_satisfaccion.describe())
```

## 5. Analisis de Ventas - Calculos Principales

```python
# Calcular ventas totales por producto y por tienda
print('VENTAS TOTALES POR TIENDA Y PRODUCTO:')

# Crear una columna de ventas totales (Cantidad_Vendida * Precio_Unitario)
df_ventas['Total_Ventas'] = df_ventas['Cantidad_Vendida'] * df_ventas['Precio_Unitario']

# Ventas totales por tienda
ventas_por_tienda = df_ventas.groupby('ID_Tienda')['Total_Ventas'].sum().sort_values(ascending=False)
print('\nVentas Totales por Tienda:')
print(ventas_por_tienda)

# Ventas totales por producto
ventas_por_producto = df_ventas.groupby('Producto')['Total_Ventas'].sum().sort_values(ascending=False)
print('\nVentas Totales por Producto:')
print(ventas_por_producto)

# Ventas totales por tienda y producto (tabla cruzada)
print('\nVentas Totales por Tienda y Producto:')
ventas_tienda_producto = df_ventas.pivot_table(values='Total_Ventas', index='ID_Tienda', columns='Producto', aggfunc='sum', fill_value=0)
print(ventas_tienda_producto)
```

```python
# Resumen estadistico detallado de ventas por tienda
print('RESUMEN ESTADISTICO DE VENTAS:')
print('-' * 50)
if 'Store_ID' in df_ventas.columns and 'Total_Ventas' in df_ventas.columns:
    resumen_ventas = df_ventas.groupby('Store_ID')['Total_Ventas'].describe()
    print(resumen_ventas)
else:
    print('No se pueden calcular estadisticas. Columnas disponibles:', df_ventas.columns.tolist())
```

## 6. Analisis de Inventarios - Rotacion y Niveles Criticos

```python
# Calcular rotacion de inventarios
print('ANALISIS DE ROTACION DE INVENTARIOS:')
print('-' * 50)

# Verificar las columnas disponibles
print('Columnas en df_inventarios:', df_inventarios.columns.tolist())
print('Columnas en df_ventas:', df_ventas.columns.tolist())

# Realizar un merge de inventarios y ventas si es necesario
# Calcular rotacion = Ventas_Totales / Stock_Disponible
# Esto dependera de la estructura real de los datos
```

```python
# Calcular porcentaje de rotacion de inventarios
print('PORCENTAJE DE ROTACION DE INVENTARIOS POR TIENDA:')
print('-' * 50)

if 'Store_ID' in df_inventarios.columns:
    # Mostrar inventario detallado por tienda
    inventario_por_tienda = df_inventarios.groupby('Store_ID').agg({
        'Stock_Disponible': 'sum',
        'Producto': 'count'  # Cantidad de productos
    }).rename(columns={'Producto': 'Num_Productos'})
    print(inventario_por_tienda)
else:
    print('Mostrando estructura de inventarios:')
    print(df_inventarios.head(10))
```

```python
# Filtrar tiendas con inventarios criticos (menos del 10%)
print('TIENDAS CON INVENTARIOS CRITICOS (< 10% de rotacion):')
print('-' * 50)

# Este filtro depended de la estructura real de los datos
# Ejemplo: df_critico = df_inventarios[df_inventarios['Porcentaje_Rotacion'] < 10]
print('Nota: El filtro de inventarios criticos se aplicara segun la estructura de datos')
```

## 7. Analisis de Satisfaccion del Cliente

```python
# Analizar satisfaccion del cliente por tienda
print('ANALISIS DE SATISFACCION DEL CLIENTE:')
print('-' * 50)

if 'Store_ID' in df_satisfaccion.columns:
    satisfaccion_por_tienda = df_satisfaccion.groupby('Store_ID')['Satisfaccion'].agg(['mean', 'min', 'max', 'count'])
    satisfaccion_por_tienda.columns = ['Satisfaccion_Promedio', 'Satisfaccion_Minima', 'Satisfaccion_Maxima', 'Num_Clientes']
    print(satisfaccion_por_tienda)
else:
    print('Mostrando estructura de satisfaccion:')
    print(df_satisfaccion.head(10))
```

```python
# Filtrar tiendas con baja satisfaccion (< 60%)
print('TIENDAS CON BAJA SATISFACCION (< 60%):')
print('-' * 50)

if 'Store_ID' in df_satisfaccion.columns and 'Satisfaccion' in df_satisfaccion.columns:
    satisfaccion_baja = df_satisfaccion[df_satisfaccion['Satisfaccion'] < 60].copy()
    
    if len(satisfaccion_baja) > 0:
        tiendas_criticas = satisfaccion_baja.groupby('Store_ID')['Satisfaccion'].agg(['mean', 'count']).sort_values('mean')
        print(tiendas_criticas)
        print('\nRECOMENDACIONES PARA MEJORAR:')
        for tienda in tiendas_criticas.index:
            promedio = tiendas_criticas.loc[tienda, 'mean']
            print(f'Tienda {tienda}: Satisfaccion promedio {promedio:.2f}% - Requiere intervencion inmediata')
    else:
        print('No hay tiendas con baja satisfaccion. Todas estan por encima del 60%')
else:
    print('Columnas necesarias no encontradas')
```

## 8. Operaciones con Numpy - Calculos Estadisticos

```python
# Calcular estadisticas de ventas usando Numpy
print('CALCULOS ESTADISTICOS CON NUMPY:')
print('-' * 50)

if 'Total_Ventas' in df_ventas.columns:
    # Convertir la columna de ventas a array de Numpy
    ventas_array = df_ventas['Total_Ventas'].to_numpy()
    
    # Calcular mediana
    mediana_ventas = np.median(ventas_array)
    print(f'Mediana de Ventas Totales: {mediana_ventas:.2f}')
    
    # Calcular desviacion estandar
    desv_estandar_ventas = np.std(ventas_array)
    print(f'Desviacion Estandar de Ventas: {desv_estandar_ventas:.2f}')
    
    # Calcular media
    media_ventas = np.mean(ventas_array)
    print(f'Media de Ventas Totales: {media_ventas:.2f}')
    
    # Calcular percentiles
    percentil_25 = np.percentile(ventas_array, 25)
    percentil_75 = np.percentile(ventas_array, 75)
    print(f'Percentil 25: {percentil_25:.2f}')
    print(f'Percentil 75: {percentil_75:.2f}')
    print(f'Rango Intercuartilico (IQR): {percentil_75 - percentil_25:.2f}')
else:
    print('Columna Total_Ventas no encontrada')
```

## 9. Proyecciones de Ventas Futuras - Simulacion con Numpy

```python
# Simular proyecciones de ventas futuras
print('SIMULACION DE PROYECCIONES DE VENTAS FUTURAS:')
print('-' * 50)

if 'Total_Ventas' in df_ventas.columns:
    # Establecer la semilla para reproducibilidad
    np.random.seed(42)
    
    # Parametros para la simulacion
    ventas_actuales = df_ventas['Total_Ventas'].to_numpy()
    media_ventas = np.mean(ventas_actuales)
    desv_estandar = np.std(ventas_actuales)
    num_meses = 12  # Proyection para 12 meses
    num_simulaciones = 1000  # Numero de simulaciones
    
    # Generar proyecciones aleatorias
    proyecciones = np.random.normal(media_ventas, desv_estandar, (num_simulaciones, num_meses))
    
    print(f'Parametros de la simulacion:')
    print(f'  Media de ventas actuales: {media_ventas:.2f}')
    print(f'  Desviacion estandar: {desv_estandar:.2f}')
    print(f'  Numero de meses proyectados: {num_meses}')
    print(f'  Numero de simulaciones: {num_simulaciones}')
    print(f'\nDimensiones del array de proyecciones: {proyecciones.shape}')
```

```python
# Analizar resultados de la simulacion
print('ANALISIS DE LAS PROYECCIONES SIMULADAS:')
print('-' * 50)

if 'Total_Ventas' in df_ventas.columns:
    # Calcular estadisticas de las proyecciones por mes
    media_proyectada = np.mean(proyecciones, axis=0)
    desv_proyectada = np.std(proyecciones, axis=0)
    minimo_proyectado = np.min(proyecciones, axis=0)
    maximo_proyectado = np.max(proyecciones, axis=0)
    
    print('\nEstadisticas de Proyecciones por Mes:')
    print('Mes | Media | Desv.Est | Minimo | Maximo')
    print('-' * 50)
    for mes in range(num_meses):
        print(f'{mes+1:3d} | {media_proyectada[mes]:8.2f} | {desv_proyectada[mes]:8.2f} | {minimo_proyectado[mes]:8.2f} | {maximo_proyectado[mes]:8.2f}')
    
    # Tendencia general
    print(f'\nVenta promedio proyectada para los 12 meses: {np.mean(media_proyectada):.2f}')
    print(f'Variabilidad esperada: {np.mean(desv_proyectada):.2f}')
```
